DATA PREPARATION

Here, the data is loaded using scikit-learn which also provides the Iris dataset in its library. The dataset is loaded to return both the feature data (X) and the labels (y) for the flowers. After which the dataset is split into training and test sets. I included stratify=y for equal class proportions. Feature scaling and feature enginnering will not be needed. 

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data     # shape (150, 4)
y = iris.target   # shape (150,)
print(iris.feature_names, iris.target_names)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

MODEL TRAINING 

Here, the model is trained using a simple Decision Tree Classifier algorithm. Using scikit-learn, the model is initailized with one hyperparameter(random_state=42). Finally the decision tree is trained with the training set. 

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=42)
```

```python
model.fit(X_train, y_train)
```

PREDICTION

After the model has been trained, it can be used to predict the species of flowers in the test set which essentially the model hasn't seen before. From the predictions, it can be understood that the model is performing well. After being able to match its predictions with the true labels. 

```python
y_pred = model.predict(X_test)
```

```python
print("Predictions:", y_pred[:5])
print("True labels:", y_test[:5])
```

EVALUATION

An evaluation is needed to assess how well the model performed. I will be making use of two evaluation metrics for this assesment; accuracy and confusion matrix. The model predicted 28/30 flowers correctly leading to an accuracy score of 0.93 (93%). The confusion matrix correctly predicted 10 true Setosa. For Versicolor, 9 were predicted correctly but 1 was mistakened for Virginica and for Virginica, 9 were predicted correctly but 1 mistakened for Versicolor. I tried experimenting with a different algorithim - k-Nearest Neighbors(k-NN). The k-NN achieved an accuracy score of 1.0 on the test set. 

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model2 = KNeighborsClassifier(n_neighbors=5)
model2.fit(X_train, y_train)
y_pred2 = model2.predict(X_test)
print("k-NN accuracy:", accuracy_score(y_test, y_pred2))
```

In this case, the model made correct species predictions for 93% of the test flowers. This is a good result leading to the conculsion that the model has successfully learned the relationships between the flower measurements and their species. However, I noticed with the use of stratify=y I was able to get a more realistic and representative evalution of the model. The stratify=y parameter ensured the train and test sets have the same class proportions as the original dataset, giving me a more honest assessment of the model's true performance. Without it, I had an accuracy score of 1.0 which gave me the impression that the model was overfitting the training data.
